```python
import numpy as np 
import pandas as pd
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
en_stopwords = stopwords.words('english')
```

```python
book_df = pd.read_csv('prog_book.csv')
book_df
```

```python
titles = book_df['Book_title'].tolist()
```

```python
filtered_tokens = []
for title in titles:
  tokens = word_tokenize(title)
  filtered_tokens.extend([token for token in tokens if (token.lower() not in en_stopwords) and (token.lower() not in string.punctuation)])
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(" ".join(filtered_tokens))
                 
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                mask=leaf_mask,
                min_font_size = 10).generate(" ".join(filtered_tokens))
                     
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
